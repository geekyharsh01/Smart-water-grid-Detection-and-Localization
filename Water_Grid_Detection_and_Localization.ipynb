```

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop('location',axis=1), df['location'], test_size=0.2, random_state=42)
```

```
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
```

```

x_train_tensor = torch.tensor(x_train.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.long)
x_test_tensor = torch.tensor(x_test.values, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.long)
```

```
import torch
import torch.nn as nn
import torch.optim as optim

l
class NN(nn.Module):
    def __init__(self):
        super(NN, self).__init__()
        self.fc1 = nn.Linear(9, 128)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 32)
        self.fc4 = nn.Linear(32,16)
        self.fc5 = nn.Linear(16,6)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        x = self.relu(x)
        x = self.fc4(x)
        x = self.relu(x)
        x = self.fc5(x)

        return x
```

```

model = NN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```
epochs = 40

for epoch in range(epochs):
  idx=0
  for i in x_train_tensor:

    optimizer.zero_grad()
    outputs = model(i)
    loss = criterion(outputs, y_train_tensor[idx])
    idx+=1
    loss.backward()
    optimizer.step()
    print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item()}')
```

```
# Model evaluation on testing data
with torch.no_grad():
    model.eval()
    predicted_labels = model(x_test_tensor).argmax(dim=1)
    accuracy = (predicted_labels == y_test_tensor).sum().item() / len(y_test)
    print(f'Training Accuracy: {accuracy * 100:.2f}%')
```

```
# Confusion matrix
cm = confusion_matrix(y_test_tensor, predicted_labels)
plt.figure(figsize=(10, 6))
sns.heatmap(cm,annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.title('Confusion Matrix')
plt.show()
```
